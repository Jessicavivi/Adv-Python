```python
#LiZiweiAsn2 - A
import sqlite3 as sql
import pandas as pd
connection = sql.Connection('dj.db')
cursor = connection.cursor()
ord1 = 'SELECT id, ticker, name FROM tickers'
cursor.execute(ord1)
rows = cursor.fetchall()
df = pd.DataFrame(rows)
df.columns = ['id','ticker','name']
df.set_index('id',inplace = True)
df
```

```python
#LiZiweiAsn2 - B

connection = sql.Connection('dj.db')
cursor = connection.cursor()
ord2 = 'SELECT min(date) as first, max(date) as last FROM daily_measures'
cursor.execute(ord2)
rows2 = cursor.fetchall()
df2 = pd.DataFrame(rows2)
df2.columns = ['earliest date','latest date']
df2.index = ['']

df2
```

```python
#LiZiweiAsn2 - C

connection = sql.Connection('dj.db')
cursor = connection.cursor()
ord3 = 'SELECT date, closing FROM daily_measures WHERE ticker_id=3 ORDER BY date'
cursor.execute(ord3)
rows3 = cursor.fetchall()
df3 = pd.DataFrame(rows3)
highest_price = df3.loc[:,1].max()
df_higPrice = df3.loc[df3[1]==highest_price]
df_higPrice.columns = ['date','price']
df_higPrice.index = ['']

df_higPrice
```

```python
#LiZiweiAsn2 - D
connection = sql.Connection('dj.db')
cursor = connection.cursor()
ord4 = 'SELECT date, closing FROM daily_measures WHERE ticker_id=3 ORDER BY date'
cursor.execute(ord4)
rows4 = cursor.fetchall()
df4 = pd.DataFrame(rows4)
df4.columns = ['date','closing']
price = df4['closing'].tolist()
price_perc_moves = []
for i in range(1,len(price)):
    percentMove = (price[i]-price[i-1])/price[i-1]
    price_perc_moves.append(percentMove)
price_perc_moves.append(0.00)
#print(len(price))
#print(len(price_perc_moves))
df4['price percentage moves'] = price_perc_moves   
df4['price percentage moves'] = df4['price percentage moves'].apply(lambda x: format(x, '.2%'))
df4       
```
