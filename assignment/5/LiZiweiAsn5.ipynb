```python
import pandas as pd
```

```python
df = pd.read_csv('DIS.csv',index_col=0, parse_dates=True)
df_500 = pd.read_csv('S&P500.csv',index_col=0, parse_dates=True)
```

```python
#B  Create a line chart for the adjusted closing prices of DIS stock
import matplotlib.pyplot as plt
ad_closing = df['Adj Close']
plt.plot(ad_closing)
plt.show()
```

```python
# C  Price Time Series  Compare 
ad_closing_500 = df_500['Adj Close']

fig, ax1 = plt.subplots() 
ad_closing_500.plot(ax=ax1, color = 'g') 
ad_closing.plot(secondary_y=True, ax=ax1) 

ax1.set_title('DIS & S&P200')
ax1.set_ylabel('S&P500')
ax1.right_ax.set_ylabel('DIS')

plt.show() 
```

# C2
The DIS trend is roughly the same as the SP500. But when SP500 showed an upward trend in 2017, DIS showed a downward trend. Overall, the performance of the DIS is basically the same as that of the SP500.

```python
#D  Analysis - Stock Beta 
def return_daily(df):
    df['daily_return'] = df['Adj Close'].pct_change()
    return df
return_daily(df)
return_daily(df_500)
```

```python
x = df_500['daily_return']
xret = x['2018-01-02':'2020-01-02']

y = df['daily_return']
yret = y['2018-01-02':'2020-01-02']

plt.title("DIS vs SP500 - Daily Returns")
plt.scatter(xret, yret)
plt.xlabel('SP500 returns')
plt.ylabel('AAPL returns')
plt.show()
```

```python
from sklearn import linear_model
df_lineR = pd.DataFrame({'SP500r':xret,'DISr':yret})

i = df_lineR[['SP500r']]
j = df_lineR['DISr']
reg = linear_model.LinearRegression()

reg.fit(i,j)

a = float(reg.coef_)
b = float(reg.intercept_)

print('equation of the regression liney = {:.2} + {:.2} * x'.format(a, b))
plt.scatter(i, j)
plt.plot(i, reg.predict(i), color='red')
plt.show()
```

```python
#D2 get linear regression report between two daily returns 
import statsmodels.api as sm

indvars = sm.add_constant(xret) 
model = sm.OLS(yret, indvars).fit()
summary = model.summary()
print(summary)
```
