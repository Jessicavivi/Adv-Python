```python
#LiZiweiAsn1-B
import xlrd
import pandas as pd
filename = 'UBER 1Y.xlsx'
wb = xlrd.open_workbook(filename)
sheet = wb.sheet_by_index(0)


#get each column in sheet
date_value = sheet.col_values(0)     
open_value = sheet.col_values(1)
high_value = sheet.col_values(2)
low_value = sheet.col_values(3)
close_value = sheet.col_values(4)
adj_close_value = sheet.col_values(5)     
volume_value = sheet.col_values(6)

#build lists for every columns
date = []        
open = []
high =[]
low = []
close = []
adj_close = []
volume = []
daily_return = []
for i in range(2,len(open_value)):
    datetime = xlrd.xldate.xldate_as_datetime(date_value[i], wb.datemode)       # get datetime 
    date.append(datetime)
    open.append(open_value[i])
    high.append(high_value[i])
    low.append(low_value[i])
    adj_close.append(adj_close_value[i])
    volume.append(volume_value[i])
    return_daily = (adj_close_value[i] - adj_close_value[i-1]) / adj_close_value[i-1]
    daily_return.append(return_daily)
df = pd.DataFrame()

# put lists in dataframe and limit the number of decimal places
df['Date'] = date                       
df['Open'] = open
df['Open'] = df['Open'].round(2)
df['High'] = high
df['High'] = df['High'].round(2)
df['Low'] = low
df['Low'] = df['Low'].round(2)
df['Adj_Close'] = adj_close
df['Adj_Close'] = df['Adj_Close'].round(2)
df['Volume'] = volume
df['Daily Return'] = daily_return
df['Daily Return'] = df['Daily Return'].apply(lambda x: format(x, '.2%'))
df

```
