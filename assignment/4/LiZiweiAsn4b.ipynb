```python
#ZiweiLi Assignment4b
import requests
import os
from dotenv import load_dotenv
```

```python
env_file = 'my_keys.env'
load_dotenv(env_file)
myAPIKey = os.getenv('ALPHAVANTAGE_KEY')
print(myAPIKey)
```

```python
# define the tickets of interest
symbols = ['KO', 'DIS']
# define the url where the web service is
API_URL = "https://www.alphavantage.co/query"
```

```python
    ko = {
        "function": "TIME_SERIES_INTRADAY",
        "symbol": 'KO',
        "interval": "5min",
        "datatype": "json",
        "apikey": myAPIKey
             }

    response = requests.get(API_URL, ko)
    ko = response.json()
    
    
    a = (ko['Time Series (5min)'])
    keys = (a.keys())
    ko_date = []
    ko_close = []
    
    for key in keys:
        ko_date.append(key)
        ko_close.append(float(a[key]['4. close']))
    print(ko_date)
    print(ko_close)
```

```python
    dis = {
        "function": "TIME_SERIES_INTRADAY",
        "symbol": 'DIS',
        "interval": "5min",
        "datatype": "json",
        "apikey": myAPIKey
             }

    response = requests.get(API_URL, dis)
    dis = response.json()
    
    
    a = (dis['Time Series (5min)'])
    keys = (a.keys())
    dis_date = []
    dis_close = []
    
    for key in keys:
        #dis_date.append(key)
        dis_close.append(float(a[key]['4. close']))
    #print(dis_date)
    print(dis_close)
```

```python
import pandas as pd
dic = {'Date':ko_date,
       'KO_CLOSE':ko_close,
       'DIS_CLOSE':dis_close}
df1 =  pd.DataFrame(dic)
df1 = df1.set_index("Date")
df1
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots() 
df1['KO_CLOSE'].plot(ax=ax1,color='g') 
df1['DIS_CLOSE'].plot(secondary_y=True, ax=ax1,color='b') 

plt.show() 
```
